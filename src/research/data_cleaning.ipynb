```python
for col in ["ProperName", "BayerFlamsteed", "Gliese"]:
    df1[col] = df1[col].astype(str).str.strip()
    df1.loc[df1[col].isin(["nan", "NaN", "None", ""]), col] = pd.NA

has_any_name = df1[["ProperName", "BayerFlamsteed", "Gliese"]].notna().any(axis=1)

no_name_count = (~has_any_name).sum()
total_count = len(df1)

print(f"Всего звёзд: {total_count}")
print(f"Без ProperName, BayerFlamsteed и Gliese: {no_name_count}")
print(f"Это {no_name_count/total_count*100:.2f}% от каталога")
```

```python
cols = ["HIP", "HD", "HR", "ProperName", "BayerFlamsteed", "Gliese"]

# На всякий случай убедимся, что типы ключей совпадают
df1["StarID"] = df1["StarID"].astype(df2["StarID"].dtype)
```

```python
df2 = df2.merge(
    df1[["StarID"] + cols],   # берем только нужные колонки
    on="StarID",
    how="left"                # не теряем строки из df2
)
```

```python
df2
```

```python
df2['HR'].dtype
```

```python
for c in ["HIP", "HD"]:
    df2[c] = pd.to_numeric(df2[c], errors="coerce")  # всё нечисловое -> NaN

# 2) Забиваем NaN нужными значениями
df2 = df2.fillna({
    "HIP": 0.0,
    "HD": 0.0,
    "HR": 0.0,
    "ProperName": "",
    "BayerFlamsteed": "",
    "Gliese": ""
})

# 3) Жёстко фиксируем типы
df2[["HIP", "HD"]] = df2[["HIP", "HD"]].astype("float64")
df2[["ProperName", "BayerFlamsteed", "Gliese"]] = (
    df2[["ProperName", "BayerFlamsteed", "Gliese"]].astype("string")  # pandas StringDtype
)
```
